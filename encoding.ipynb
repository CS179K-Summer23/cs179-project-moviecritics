```python
import pandas as pd
df = pd.read_csv("clean_data.csv")
```

```python
df[['title','genres']].head(50)
```

```python
# Specific string to search for
#specific_string = 'Comedy'

# User input for the specific string to search for
specific_string = input("Enter the specific string to search for: ")


# Boolean indexing to get rows with the specific string in the column
list_specific_string = df[df['genres'] == specific_string]

# Print the resulting DataFrame
#print(list_specific_string[['title','genres']])
list_specific_string[['title','genres']]
```

```python
# User input for the specific genre to search for
specific_genre = input("Enter the specific genre to search for: ")

# Boolean indexing to get rows with the specific genre in the column (case-insensitive)
matching_rows = df[df['genres'].str.lower().str.contains(specific_genre, na=False)]

# Check if there are any rows with the specific genre
if not matching_rows.empty:
    # Print only the 'title' and 'column_name' columns for the rows with the specific genre
    print(matching_rows[['title', 'genres']])
else:
    print("No rows found with the specific genre.")
```

```python
# Split the 'genres' column based on hyphens, then explode the list-like values into separate rows
df_exploded = df['genres'].str.split('-', expand=True).stack().reset_index(level=1, drop=True)

# Drop duplicates to get unique strings
unique_genres = df_exploded.drop_duplicates()

# Print the unique genres
print(unique_genres)
```

```python
# Split the 'genres' column based on hyphens, then get dummies for each genre
df_encoded = df['genres'].str.get_dummies('-')

# Print the encoded DataFrame
df_encoded 
```
